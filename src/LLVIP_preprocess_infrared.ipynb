```python
# TRAINING (images, labels)
import os
%cd /kaggle/working/dataset/train

train_img = sorted(os.listdir('/kaggle/input/llvip-dataset/LLVIP/infrared/train'))
# train_lb = sorted(os.listdir('/kaggle/working/yolov8/data/train/labels'))

# print('train_img = ',train_img)
# print('train_lb = ', train_lb)

print(len(train_img))
```

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
%cd /kaggle/working/dataset/train/images 
root = '/kaggle/input/llvip-dataset/LLVIP/infrared/train'
from tqdm import tqdm
for path in tqdm(train_img):
    img = cv.imread(os.path.join(root, path), cv.IMREAD_GRAYSCALE)
    
#     # create a CLAHE object (Arguments are optional).
#     clahe = cv.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
#     cl1 = clahe.apply(img)
    
    equ = cv.equalizeHist(img)
#     res = np.hstack((img,equ)) #stacking images side-by-side

    cv.imwrite(path,equ)
```
